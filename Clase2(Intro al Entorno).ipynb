# Listas

```python
juegos = ['Halo', 'Minecraft', 'Forza Horizon', 'Roblox']
lista_mix = [1, 'Mario', 2, 'Rayman']

print(juegos)
print(lista_mix)
print(f"El primer elemento de la lista de juegos es: {juegos[0]}")

juegos.append('Sonic')
print(juegos)
```

# Bucles

```python
contador = 0

while contador <= 5:
  print (contador)
  contador += 1

for contador in range (6):
  print(contador)

for contador in range (5, 11):
  print(contador)

for contador in range (0, 11, 2):
  print(contador)

for juego in juegos:
  print(juego)
```

# Diccionarios y Tuplas

```python
diccionario = {'gato' : 'cat',
               'perro': 'dog'}
print(diccionario['gato'])

for key in diccionario.keys():
  print(key)

for value in diccionario.values():
  print(value)
```

# Funciones

```python
def suma(a,b):
  return a+b

print (suma(2,4))
```

```python
def contar():
  contador = 0
  while contador <= 10:
    print(contador)
    contador+=1

  for i in range(0,10):
    print(i)

contar()
```

```python
def multiplicar (a,b):
  print(a*b)
multiplicar(15, 24)
```

## Archivos

```python
# Leer archivo
with open('data.txt', 'r') as file:
  for line in file:
    print(line)
```

```python
#Editar el archivo
with open('data.txt', 'a') as file:
  file.write('\nGears of War')
```

```python
# Guardar en un nuevo archivo archivo
with open('data2.txt', 'w') as file:
  file.write('Clash Royale')
```

## Excepciones

```python
try:
  dividendo = float(input('Introduce el dividendo'))
  divisor = float(input('Introduce el divisor'))
  resultado = dividendo/divisor
except Exception as e:
  print(f"Error: {e}")
else:
  print("El resultado es", resultado)
finally:
  print("Fin del programa")
```
