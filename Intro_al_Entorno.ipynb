# Listas

```python
juegos = ['Halo', 'Minecraft', 'Forza Horizon', 'Roblox']
lista_mix = [1, 'Mario', 2, 'Rayman']

print(juegos)
print(lista_mix)
print(f"El primer elemento de la lista de juegos es: {juegos[0]}")

juegos.append('Sonic')
print(juegos)
```

# Bucles

```python
contador = 0

while contador <= 5:
  print (contador)
  contador += 1

for contador in range (6):
  print(contador)

for contador in range (5, 11):
  print(contador)

for contador in range (0, 11, 2):
  print(contador)

for juego in juegos:
  print(juego)
```

# Diccionarios y Tuplas

```python
diccionario = {'gato' : 'cat',
               'perro': 'dog'}
print(diccionario['gato'])

for key in diccionario.keys():
  print(key)

for value in diccionario.values():
  print(value)
```
